```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr
import statsmodels.api as sm
from statsmodels.stats.multitest import multipletests
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
butyrate_data = pd.read_excel("data/butyrate_pick.xlsx", index_col=0)
food_data_spear = pd.read_csv('data/Summarized_Food_Categories_Only.csv', index_col=0)
participants_data = pd.read_csv('data/demo_535.csv', index_col=0)
```

```python
participants_data
```

```python
common_indices = butyrate_data.index.intersection(food_data_spear.index)
butyrate_data = butyrate_data.loc[common_indices]
food_data_spear = food_data_spear.loc[common_indices]
participants_data = participants_data.loc[common_indices]
```

```python
# Convert data to numeric and handle non-numeric entries
butyrate_data = butyrate_data.apply(pd.to_numeric, errors='coerce')
food_data_spear = food_data_spear.apply(pd.to_numeric, errors='coerce')
participants_data['BMI'] = pd.to_numeric(participants_data['BMI'], errors='coerce')
participants_data['FREG8_Age'] = pd.to_numeric(participants_data['FREG8_Age'], errors='coerce')
```

```python
def adjust_for_covariates(data, covariates, categorical_vars=[]):
    # Generate dummy variables for categorical variables
    if categorical_vars:
        covariates = pd.get_dummies(covariates, columns=categorical_vars, drop_first=True)
    
    adjusted_data = pd.DataFrame(index=data.index, columns=data.columns)
    covariates = sm.add_constant(covariates)  # Add a constant to the model
    
    for column in data.columns:
        # Drop rows where NaN exists in the current column or in covariates
        combined = pd.concat([data[column], covariates], axis=1).dropna()
        model = sm.OLS(combined[column], combined[covariates.columns]).fit()
        adjusted_data.loc[combined.index, column] = model.resid

    return adjusted_data

# Adjust microbiome and diet data
# Ensure that 'FREG5_Ethnic_Group' is included as a categorical variable
adjusted_butyrate_data = adjust_for_covariates(butyrate_data, participants_data[['BMI', 'FREG8_Age', 'FREG5_Ethnic_Group']], categorical_vars=['FREG5_Ethnic_Group'])
adjusted_food_spear = adjust_for_covariates(food_data_spear, participants_data[['BMI', 'FREG8_Age', 'FREG5_Ethnic_Group']], categorical_vars=['FREG5_Ethnic_Group'])
```

```python
adjusted_butyrate_data
```

```python
def adjust_for_covariates_old(data, covariates):
    adjusted_data = pd.DataFrame(index=data.index, columns=data.columns)
    covariates = sm.add_constant(covariates)  # Add a constant to the model
    for column in data.columns:
        # Drop rows where NaN exists in the current column or in covariates
        combined = pd.concat([data[column], covariates], axis=1).dropna()
        model = sm.OLS(combined[column], combined[covariates.columns]).fit()
        adjusted_data.loc[combined.index, column] = model.resid
    return adjusted_data



# Adjust microbiome and diet data
adjusted_butyrate_data_old = adjust_for_covariates_old(butyrate_data, participants_data[['BMI', 'FREG8_Age']])
```

```python
correlations_butyrate = pd.DataFrame(index=adjusted_butyrate_data.columns, columns=adjusted_food_spear.columns)
p_values_butyrate = pd.DataFrame(index=adjusted_butyrate_data.columns, columns=adjusted_food_spear.columns)


# Ensure the data is numeric and handle NaNs appropriately
correlations_butyrate = correlations_butyrate.astype(float)  # Convert to float
p_values_butyrate = p_values_butyrate.astype(float)  # Convert to float

# Example to handle NaNs in correlations if necessary
correlations_butyrate.fillna(0, inplace=True)  # Replace NaNs with zero or some other appropriate value
```

```python
for microbe in adjusted_butyrate_data.columns:
    for nutrient in adjusted_food_spear.columns:
        # Drop NaN values from both columns before correlation
        valid_indices = adjusted_butyrate_data[microbe].notna() & adjusted_food_spear[nutrient].notna()
        if valid_indices.any():  # Check if there are any valid pairs left
            corr, pval = spearmanr(adjusted_butyrate_data[microbe][valid_indices], adjusted_food_spear[nutrient][valid_indices])
            correlations_butyrate.loc[microbe, nutrient] = corr
            p_values_butyrate.loc[microbe, nutrient] = pval
        else:
            correlations_butyrate.loc[microbe, nutrient] = np.nan
            p_values_butyrate.loc[microbe, nutrient] = np.nan
```

```python
p_flat = p_values_butyrate.values.flatten()
reject, corrected_pvals_butyrate, _, _ = multipletests(p_flat, alpha=0.2, method='fdr_bh')
significant_matrix_butyrate = pd.DataFrame(reject.reshape(p_values_butyrate.shape), index=adjusted_butyrate_data.columns, columns=adjusted_food_spear.columns)

# Summing the significant correlations per microbe
true_counts = significant_matrix_butyrate.sum(axis=1)

# Filter species with at least 10 significant correlations
significant_species = true_counts >= 0
filtered_species_butyrate = correlations_butyrate.loc[significant_species.index[significant_species]]
```

```python
filtered_significance_matrix_butyrate = significant_matrix_butyrate.loc[filtered_species_butyrate.index, filtered_species_butyrate.columns]
clipped_correlations_butyrate = np.clip(filtered_species_butyrate, -0.15, 0.15)


plt.figure(figsize=(20, 10))
ax = sns.clustermap(clipped_correlations_butyrate, cmap="RdBu_r", annot=False, cbar_kws={'label': 'Correlation'}, vmin=-0.15, vmax=0.15, row_cluster=True, col_cluster=True)

ax.ax_row_dendrogram.set_visible(False)
ax.ax_col_dendrogram.set_visible(False)

row_order = ax.dendrogram_row.reordered_ind
col_order = ax.dendrogram_col.reordered_ind

for (i, j), significant in np.ndenumerate(filtered_significance_matrix_butyrate):
    if significant:
        # Find the new position in the clustered heatmap
        new_i = row_order.index(i)
        new_j = col_order.index(j)
        ax.ax_heatmap.text(new_j + 0.5, new_i + 0.5, '*', ha='center', va='center', color='black', size=10)




#for (i, j), significant in np.ndenumerate(filtered_significance_matrix_spear):
    #if significant:
        #ax.text(j + 0.5, i + 0.5, '*', ha='center', va='center', color='black', size=10)
      
#plt.title('Adjusted Spearman Correlations between Microbial Species and Nutrition')
#plt.xlabel('Food Gr')
#plt.ylabel('Spearman correlation')

plt.rc('font', size=15)
plt.rc('axes', labelsize=15)
plt.rc('axes', titlesize=15)
#plt.rc('figure', titlesize=)
plt.show()

#plt.savefig('Repeat_Predictnutrigroup_only.png', dpi=300)  


#plt.close()
```

```python

race_data = pd.get_dummies(participants_data['FREG5_Ethnic_Group'])

# Compute correlations and p-values
correlations_race = pd.DataFrame(index=race_data.columns, columns=adjusted_butyrate_data.columns)
p_values_race = pd.DataFrame(index=race_data.columns, columns=adjusted_butyrate_data.columns)

for race in race_data.columns:
    for microbe in adjusted_butyrate_data.columns:
        valid_indices = race_data[race].notna() & adjusted_butyrate_data[microbe].notna()
        if valid_indices.any():
            corr, pval = spearmanr(race_data[race][valid_indices], adjusted_butyrate_data[microbe][valid_indices])
            correlations_race.loc[race, microbe] = corr
            p_values_race.loc[race, microbe] = pval
        else:
            correlations_race.loc[race, microbe] = np.nan
            p_values_race.loc[race, microbe] = np.nan

# Apply multiple testing correction
p_flat = p_values_race.values.flatten()
reject, corrected_pvals_race, _, _ = multipletests(p_flat, alpha=0.05, method='fdr_bh')  # Change alpha as necessary
significant_matrix_race = pd.DataFrame(reject.reshape(p_values_race.shape), index=race_data.columns, columns=adjusted_butyrate_data.columns)

# Clipping correlation values for visualization
clipped_correlations_race = np.clip(correlations_race.astype(float), -0.15, 0.15)

# Visualization
plt.figure(figsize=(20, 10))
ax = sns.clustermap(clipped_correlations_race, cmap="RdBu_r", annot=False, cbar_kws={'label': 'Correlation'}, vmin=-0.15, vmax=0.15, row_cluster=True, col_cluster=True)
ax.ax_row_dendrogram.set_visible(False)
ax.ax_col_dendrogram.set_visible(False)

row_order = ax.dendrogram_row.reordered_ind
col_order = ax.dendrogram_col.reordered_ind

# Marking significant correlations with a '*'
for (i, j), significant in np.ndenumerate(significant_matrix_race):
    if significant:
        new_i = row_order.index(i)
        new_j = col_order.index(j)
        ax.ax_heatmap.text(new_j + 0.5, new_i + 0.5, '*', ha='center', va='center', color='black', size=10)

plt.rc('font', size=15)
plt.rc('axes', labelsize=15)
plt.rc('axes', titlesize=15)
plt.show()
```

```python

correlations_race_food = pd.DataFrame(index=race_data.columns, columns=adjusted_food_spear.columns)
p_values_race_food = pd.DataFrame(index=race_data.columns, columns=adjusted_food_spear.columns)

for race in race_data.columns:
    for food in adjusted_food_spear.columns:
        valid_indices = race_data[race].notna() & adjusted_food_spear[food].notna()
        if valid_indices.any():
            corr, pval = spearmanr(race_data[race][valid_indices], adjusted_food_spear[food][valid_indices])
            correlations_race_food.loc[race, food] = corr
            p_values_race_food.loc[race, food] = pval
        else:
            correlations_race_food.loc[race, food] = np.nan
            p_values_race_food.loc[race, food] = np.nan

# Apply multiple testing correction
p_flat = p_values_race_food.values.flatten()
reject, corrected_pvals_race_food, _, _ = multipletests(p_flat, alpha=0.05, method='fdr_bh')  # Adjust alpha as necessary
significant_matrix_race_food = pd.DataFrame(reject.reshape(p_values_race_food.shape), index=race_data.columns, columns=adjusted_food_spear.columns)

# Clipping correlation values for visualization
clipped_correlations_race_food = np.clip(correlations_race_food.astype(float), -0.15, 0.15)

# Visualization
plt.figure(figsize=(20, 10))
ax = sns.clustermap(clipped_correlations_race_food, cmap="RdBu_r", annot=False, cbar_kws={'label': 'Correlation'}, vmin=-0.15, vmax=0.15, row_cluster=True, col_cluster=True)
ax.ax_row_dendrogram.set_visible(False)
ax.ax_col_dendrogram.set_visible(False)

row_order = ax.dendrogram_row.reordered_ind
col_order = ax.dendrogram_col.reordered_ind

# Marking significant correlations with a '*'
for (i, j), significant in np.ndenumerate(significant_matrix_race_food):
    if significant:
        new_i = row_order.index(i)
        new_j = col_order.index(j)
        ax.ax_heatmap.text(new_j + 0.5, new_i + 0.5, '*', ha='center', va='center', color='black', size=10)

plt.rc('font', size=15)
plt.rc('axes', labelsize=15)
plt.rc('axes', titlesize=15)
plt.show()
```

```python

# Convert all data to numeric, assuming that 'race_data' needs to be coerced into numeric types
race_data = race_data.apply(pd.to_numeric, errors='coerce')
food_data_spear = food_data_spear.apply(pd.to_numeric, errors='coerce')

# Drop any remaining NaN values (if any) or fill them
race_data.fillna(0, inplace=True)
food_data_spear.fillna(0, inplace=True)

# Calculate the sum or mean of food consumption per race
abundance_data = pd.DataFrame(index=race_data.columns, columns=food_data_spear.columns)
for race in race_data.columns:
    for food in food_data_spear.columns:
        # Calculate the sum of food consumption weighted by race indicator
        abundance_data.loc[race, food] = (race_data[race] * food_data_spear[food]).sum()

# Optionally, normalize by the total consumption of each race
abundance_data_normalized = abundance_data.div(abundance_data.sum(axis=1), axis=0)

# Visualization of normalized data
plt.figure(figsize=(12, 8))
sns.heatmap(abundance_data, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Normalized Abundance'})
plt.title('Normalized Abundance of Food Consumption by Race')
plt.xlabel('Food Types')
plt.ylabel('Races')
plt.show()
```
