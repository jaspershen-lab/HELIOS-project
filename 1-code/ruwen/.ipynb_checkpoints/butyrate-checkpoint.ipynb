```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Load the Excel file
file_path = 'data/species_535.xlsx'  # Change this to your file's path
data = pd.read_excel(file_path)

# Load the workbook and worksheet
workbook = load_workbook(file_path)
worksheet = workbook.active

# Define your categories (make sure these are correct as per your data)
#bacteria = ['list_of_bacteria_species']  # Replace with actual names or load from a file
#viruses = ['list_of_virus_species']  # Replace with actual names or load from a file

# Set the fill colors
#bacteria_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
virus_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

# Apply the fills based on categories
for col in worksheet.iter_cols(min_row=1, max_row=1, min_col=2, max_col=len(data.columns)):
    cell = col[0]
    if cell.value in viruses:
        for c in col:
            c.fill = virus_fill

# Save the modified file
workbook.save('highlighted_species_535.xlsx')  # Change this to your desired output path
```

```python
butyrate_data = pd.read_excel("data/butyrate_producing_species_18.xlsx", index_col=0)
food_data_spear = pd.read_csv('data/Summarized_Food_Categories_Only.csv', index_col=0)
participants_data = pd.read_csv('data/demo_535.csv', index_col=0)
```

butyrate_data

```python
butyrate_data
```

```python
common_indices = butyrate_data.index.intersection(food_data_spear.index)
butyrate_data = butyrate_data.loc[common_indices]
food_data_spear = food_data_spear.loc[common_indices]
participants_data = participants_data.loc[common_indices]
```

```python
# Convert data to numeric and handle non-numeric entries
butyrate_data = butyrate_data.apply(pd.to_numeric, errors='coerce')
food_data_spear = food_data_spear.apply(pd.to_numeric, errors='coerce')
participants_data['BMI'] = pd.to_numeric(participants_data['BMI'], errors='coerce')
participants_data['FREG8_Age'] = pd.to_numeric(participants_data['FREG8_Age'], errors='coerce')
```

```python
def adjust_for_covariates(data, covariates):
    adjusted_data = pd.DataFrame(index=data.index, columns=data.columns)
    covariates = sm.add_constant(covariates)  # Add a constant to the model
    for column in data.columns:
        # Drop rows where NaN exists in the current column or in covariates
        combined = pd.concat([data[column], covariates], axis=1).dropna()
        model = sm.OLS(combined[column], combined[covariates.columns]).fit()
        adjusted_data.loc[combined.index, column] = model.resid
    return adjusted_data



# Adjust microbiome and diet data
adjusted_butyrate_data = adjust_for_covariates(butyrate_data, participants_data[['BMI', 'FREG8_Age']])
adjusted_food_spear = adjust_for_covariates(food_data_spear, participants_data[['BMI', 'FREG8_Age']])
```

```python
correlations_butyrate = pd.DataFrame(index=adjusted_butyrate_data.columns, columns=adjusted_food_spear.columns)
p_values_butyrate = pd.DataFrame(index=adjusted_butyrate_data.columns, columns=adjusted_food_spear.columns)


# Ensure the data is numeric and handle NaNs appropriately
correlations_butyrate = correlations_butyrate.astype(float)  # Convert to float
p_values_butyrate = p_values_butyrate.astype(float)  # Convert to float

# Example to handle NaNs in correlations if necessary
correlations_butyrate.fillna(0, inplace=True)  # Replace NaNs with zero or some other appropriate value
```

```python
for microbe in adjusted_butyrate_data.columns:
    for nutrient in adjusted_food_spear.columns:
        # Drop NaN values from both columns before correlation
        valid_indices = adjusted_butyrate_data[microbe].notna() & adjusted_food_spear[nutrient].notna()
        if valid_indices.any():  # Check if there are any valid pairs left
            corr, pval = spearmanr(adjusted_butyrate_data[microbe][valid_indices], adjusted_food_spear[nutrient][valid_indices])
            correlations_butyrate.loc[microbe, nutrient] = corr
            p_values_butyrate.loc[microbe, nutrient] = pval
        else:
            correlations_butyrate.loc[microbe, nutrient] = np.nan
            p_values_butyrate.loc[microbe, nutrient] = np.nan

#significant_matrix_spear, corrected_pvals_spear, _, _ = multipletests(p_values_spear.values.flatten(), alpha=0.2, method='fdr_bh')

#significant_matrix_spear = significant_matrix_spear.reshape(p_values_spear.shape)

```

```python
p_flat = p_values_butyrate.values.flatten()
reject, corrected_pvals_butyrate, _, _ = multipletests(p_flat, alpha=0.2, method='fdr_bh')
significant_matrix_butyrate = pd.DataFrame(reject.reshape(p_values_butyrate.shape), index=adjusted_butyrate_data.columns, columns=adjusted_food_spear.columns)

# Summing the significant correlations per microbe
true_counts = significant_matrix_butyrate.sum(axis=1)

# Filter species with at least 10 significant correlations
significant_species = true_counts >= 1
filtered_species_butyrate = correlations_butyrate.loc[significant_species.index[significant_species]]
```

```python
filtered_species_butyrate
```

```python
filtered_significance_matrix_butyrate = significant_matrix_butyrate.loc[filtered_species_butyrate.index, filtered_species_butyrate.columns]
clipped_correlations_butyrate = np.clip(filtered_species_butyrate, -0.15, 0.15)


plt.figure(figsize=(20, 10))
ax = sns.clustermap(clipped_correlations_butyrate, cmap="RdBu_r", annot=False, cbar_kws={'label': 'Correlation'}, vmin=-0.15, vmax=0.15, row_cluster=True, col_cluster=True)

ax.ax_row_dendrogram.set_visible(False)
ax.ax_col_dendrogram.set_visible(False)

row_order = ax.dendrogram_row.reordered_ind
col_order = ax.dendrogram_col.reordered_ind

for (i, j), significant in np.ndenumerate(filtered_significance_matrix_butyrate):
    if significant:
        # Find the new position in the clustered heatmap
        new_i = row_order.index(i)
        new_j = col_order.index(j)
        ax.ax_heatmap.text(new_j + 0.5, new_i + 0.5, '*', ha='center', va='center', color='black', size=10)




#for (i, j), significant in np.ndenumerate(filtered_significance_matrix_spear):
    #if significant:
        #ax.text(j + 0.5, i + 0.5, '*', ha='center', va='center', color='black', size=10)
      
#plt.title('Adjusted Spearman Correlations between Microbial Species and Nutrition')
#plt.xlabel('Food Gr')
#plt.ylabel('Spearman correlation')

plt.rc('font', size=15)
plt.rc('axes', labelsize=15)
plt.rc('axes', titlesize=15)
#plt.rc('figure', titlesize=)
plt.show()

#plt.savefig('Repeat_Predictnutrigroup_only.png', dpi=300)  


#plt.close()
```
