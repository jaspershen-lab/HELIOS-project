```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr
import statsmodels.api as sm
from statsmodels.stats.multitest import multipletests
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


# Example Data Load
species_data = pd.read_csv('species_predict.csv', index_col=0)

total_counts = species_data.sum(axis=1)

# Divide each count by the row sum to get relative abundances
species_data = species_data.div(total_counts, axis=0)

diet_data = pd.read_csv('predict_food_group.csv', index_col=0)
participants_data = pd.read_csv('demo_535.csv', index_col=0)
```

```python
common_indices = species_data.index.intersection(diet_data.index)
species_data = species_data.loc[common_indices]
diet_data = diet_data.loc[common_indices]
participants_data = participants_data.loc[common_indices]
```

```python
# Convert data to numeric and handle non-numeric entries
species_data = species_data.apply(pd.to_numeric, errors='coerce')
diet_data = diet_data.apply(pd.to_numeric, errors='coerce')
participants_data['BMI'] = pd.to_numeric(participants_data['BMI'], errors='coerce')
participants_data['FREG8_Age'] = pd.to_numeric(participants_data['FREG8_Age'], errors='coerce')

# Optionally, fill NaN values or drop them
#microbiome_data = microbiome_data.fillna(microbiome_data.mean())  # Or use another method to handle NaNs
#diet_data = diet_data.fillna(diet_data.mean())
```

```python
# Adjust for BMI and Age



def adjust_for_covariates(data, covariates):
    adjusted_data = pd.DataFrame(index=data.index, columns=data.columns)
    covariates = sm.add_constant(covariates)  # Add a constant to the model
    for column in data.columns:
        # Drop rows where NaN exists in the current column or in covariates
        combined = pd.concat([data[column], covariates], axis=1).dropna()
        model = sm.OLS(combined[column], combined[covariates.columns]).fit()
        adjusted_data.loc[combined.index, column] = model.resid
    return adjusted_data



# Adjust microbiome and diet data
adjusted_microbiome = adjust_for_covariates(species_data, participants_data[['BMI', 'FREG8_Age']])
adjusted_diet = adjust_for_covariates(diet_data, participants_data[['BMI', 'FREG8_Age']])

```

```python
adjusted_microbiome
```

```python
# Compute Spearman correlations on adjusted data

correlations = pd.DataFrame(index=adjusted_microbiome.columns, columns=adjusted_diet.columns)
p_values = pd.DataFrame(index=adjusted_microbiome.columns, columns=adjusted_diet.columns)


# Ensure the data is numeric and handle NaNs appropriately
correlations = correlations.astype(float)  # Convert to float
p_values = p_values.astype(float)  # Convert to float

# Example to handle NaNs in correlations if necessary
correlations.fillna(0, inplace=True)  # Replace NaNs with zero or some other appropriate value


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check the type of data being passed to heatmap
#print(correlations.dtypes)  # Ensure all data is float
#print(correlations.isna().any())  # Check for any remaining NaNs

# Visualize the heatmap
#plt.figure(figsize=(20, 10))
#ax = sns.heatmap(correlations, annot=False, cmap='coolwarm', cbar_kws={'label': 'Adjusted Spearman Correlation'})
#plt.show()
```

```python
for microbe in adjusted_microbiome.columns:
    for nutrient in adjusted_diet.columns:
        # Drop NaN values from both columns before correlation
        valid_indices = adjusted_microbiome[microbe].notna() & adjusted_diet[nutrient].notna()
        if valid_indices.any():  # Check if there are any valid pairs left
            corr, pval = spearmanr(adjusted_microbiome[microbe][valid_indices], adjusted_diet[nutrient][valid_indices])
            correlations.loc[microbe, nutrient] = corr
            p_values.loc[microbe, nutrient] = pval
        else:
            correlations.loc[microbe, nutrient] = np.nan
            p_values.loc[microbe, nutrient] = np.nan

significant_matrix, corrected_pvals, _, _ = multipletests(p_values.values.flatten(), alpha=0.2, method='fdr_bh')

significant_matrix = significant_matrix.reshape(p_values.shape)

plt.figure(figsize=(20, 10))
ax = sns.heatmap(correlations, annot=False, cmap='coolwarm', cbar_kws={'label': 'Adjusted Spearman Correlation'})
for (i, j), significant in np.ndenumerate(significant_matrix):
    if significant:
        ax.text(j + 0.5, i + 0.5, '*', ha='center', va='center', color='black')

plt.title('Adjusted Spearman Correlations between Microbial Species and Dietary Components')
plt.xlabel('Dietary Components')
plt.ylabel('Microbial Species')
plt.show()
```

```python
def adjust_for_covariates(data, covariates):
    adjusted_data = pd.DataFrame(index=data.index, columns=data.columns)
    covariates = sm.add_constant(covariates)  # Add a constant to the model
    for column in data.columns:
        # Drop rows where NaN exists in the current column or in covariates
        combined = pd.concat([data[column], covariates], axis=1).dropna()
        model = sm.OLS(combined[column], combined[covariates.columns]).fit()
        adjusted_data.loc[combined.index, column] = model.resid
    return adjusted_data



# Adjust microbiome and diet data
adjusted_microbiome_spear = adjust_for_covariates(microbiome_data_spear, participants_data[['BMI', 'FREG8_Age']])
adjusted_food_spear = adjust_for_covariates(food_data_spear, participants_data[['BMI', 'FREG8_Age']])
```
