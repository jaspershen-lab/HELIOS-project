```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python

data = {
    'source': ['food1', 'food1', 'nutrient1', 'food2', 'nutrient2'],
    'target': ['nutrient1', 'nutrient2', 'species1', 'species1', 'species2'],
    'weight': [1, 0.5, 2, 1.5, 3]
}

df = pd.DataFrame(data)

# Create network
G = nx.from_pandas_edgelist(df, 'source', 'target', edge_attr=True)

# Visualize the network
pos = nx.spring_layout(G, seed=42)  # This uses a fixed seed for layout consistency
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='gray', width=4)
plt.show()
```
